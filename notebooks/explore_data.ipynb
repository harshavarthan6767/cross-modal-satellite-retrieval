# Explore the SEN12MS triplet data

Visualise sample SAR / optical / multispectral triplets and their land-cover labels to sanity-check the pipeline before training.

Run from the repo root after preparing data:

```
python data/download_sen12ms.py --synthetic --num 200 --out data/sen12ms
```

````python
import sys, pathlib
sys.path.insert(0, str(pathlib.Path.cwd()))
import numpy as np, pandas as pd, matplotlib.pyplot as plt
from config import load_config
cfg = load_config()
manifest = pd.read_parquet(pathlib.Path(cfg['paths']['data_root']) / 'manifest.parquet')
print('triplets:', len(manifest))
manifest.head()
````

````python
ax = manifest['label'].value_counts().sort_index().plot.bar()
ax.set_xlabel('land-cover class'); ax.set_ylabel('count'); ax.set_title('Class balance')
````

````python
def show_triplet(row):
    sar = np.load(row['sar_path'])
    opt = np.load(row['opt_path'])
    ms  = np.load(row['ms_path'])
    fig, axes = plt.subplots(1, 3, figsize=(9, 3))
    axes[0].imshow(sar[0], cmap='gray'); axes[0].set_title(f"SAR VV\n{row['class_name']}")
    axes[1].imshow(np.transpose(opt, (1,2,0))); axes[1].set_title('Optical RGB')
    # multispectral false-colour: NIR,R,G
    axes[2].imshow(np.transpose(ms[[3,2,0]], (1,2,0))); axes[2].set_title('MS (NIR,R,G)')
    for a in axes: a.axis('off')
    plt.tight_layout(); plt.show()

for i in range(3):
    show_triplet(manifest.iloc[i])
````
